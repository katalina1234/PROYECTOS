```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
class ACP:

 def __init__(self, df, col_string = True, X_scale = True):
    self.df = df
    self.col_string = col_string
    self.X_scale = X_scale
    self.eigvecs = None
    
 def string(self):
   if self.col_string == True:
     self.df = pd.DataFrame(self.df)
     for i in range(self.df.shape[1]):
       x = (self.df.iloc[:,[i]].dtypes == 'O')
       x = x.bool()

       if x == True:
         self.df = np.delete(np.array(self.df), i, axis=1)
         return self.df  
            
   else:
     self.df = pd.DataFrame(self.df)
     self.df = self.df.values
     return self.df
            
 def inercia(self):
   if self.X_scale == True:
     X_std = StandardScaler().fit_transform(self.df)
     self.df = X_std
   else:
     self.df = self.df  

   cov = self.df.T.dot(self.df)/self.df.shape[0]
   eigvals, self.eigvecs = np.linalg.eig(cov)
   varianza_explicada = {i:i/sum(eigvals) for i in eigvals}

   iner = pd.DataFrame({'eigvalues': eigvals,
                        'inercia_acum': np.cumsum(eigvals),
                        'Var_expl': varianza_explicada.values(),
                        'Var_expl_acum':np.cumsum([i for i in varianza_explicada.values()])}) 
   return iner

 def proyec(self):
     for i in range(self.df.shape[1]):
       p = pd.DataFrame((np.dot(self.eigvecs.T, self.df.T)).T)
     for j in range(self.df.shape[1]): 
       p = p.rename(columns = {j : f'CP{[j]}'})
     return p
```
